# Ames Housing Project Suggestions

Data science is not a linear process. In this project, in particular, you will likely find that EDA, data cleaning, and exploratory visualizations will constantly feed back into each other. Here's an example:

1. During basic EDA, you identify many missing values in a column/feature.
2. You consult the data dictionary and use domain knowledge to decide _what_ is meant by this missing feature.
3. You impute a reasonable value for the missing value.
4. You plot the distribution of your feature.
5. You realize what you imputed has negatively impacted your data quality.
6. You cycle back, re-load your clean data, re-think your approach, and find a better solution.

Then you move on to your next feature. _There are dozens of features in this dataset._

Figuring out programmatically concise and repeatable ways to clean and explore your data will save you a lot of time.

The outline below does not necessarily cover every single thing that you will want to do in your project. You may choose to do some things in a slightly different order. Many students choose to work in a single notebook for this project. Others choose to separate sections out into separate notebooks. Check with your local instructor for their preference and further suggestions.

## EDA
- **Read the data dictionary.**
- Determine _what_ missing values mean.
- Figure out what each categorical value represents.
- Identify outliers.
- Consider whether discrete values are better represented as categorical or continuous.

## Data Cleaning
- Decide how to impute null values.
- Decide how to handle outliers.
- Do you want to combine any features?
- Do you want to have interaction terms? 
- Do you want to manually drop collinear features?

## Exploratory Visualizations
- Look at distributions.
- Look at correlations.
- Look at relationships to target.

## Pre-processing
- One-hot encode categorical variables.
- Train/test split your data.
- Scale your data.
- Consider using automated feature selection.

## Modeling
- **Establish your baseline score.**
- Fit linear regression. Look at your coefficients. Are any of them wildly overblown?
- Fit lasso/ridge/elastic net with default parameters.
- Go back and remove features that might be causing issues in your models.
- Tune hyperparameters.
- **Identify a production model.** (This does not have to be your best performing Kaggle model, but rather the model that best answers your problem statement.)
- Refine and interpret your production model.

## Inferential Visualizations
- Look at feature loadings.
- Look at how accurate your predictions are.
- Is there a pattern to your errors? Consider reworking your model to address this.

## Business Recommendations
- Which features appear to add the most value to a home?
- Which features hurt the value of a home the most?
- What are things that homeowners could improve in their homes to increase the value?
- What neighborhoods seem like they might be a good investment?
- Do you feel that this model will generalize to other cities? How could you revise your model to make it more universal OR what date would you need from another city to make a comparable model?

### 8.. Handling Text and Categorical Attributes

We have chosen our categorical variables above so now we we will encode it into binary columns. `get_dummies` is preferred because i had an issue with hot encoder in that i could not get the column names back. After that we will introduct polynomial features and scale the data.

```python
housing_cat_train = pd.get_dummies(train.loc[:, ['neighborhood', 'condition_2', 'bsmt_qual']])
```

### 9. Preprocessing

Introduce Polynomial features and scaling of numerical data before merging in single dataframe Need to ensure both data has similar rows. Point to note, it is always ideal to not drop any rows unnecessarily.

```python
#import and specifying the numerical columns that were chosen
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
housing_tr1 = housing_tr1[['overall_qual', 'year_built', 'gr_liv_area', 'total_bsmt_sf', 'garage_area']]
```

```python
#Introducing polynomial features
poly = PolynomialFeatures()
X_poly = poly.fit_transform(housing_tr1)
```

```python
#Converting the polynomials into a dataframe with the appropriate columns names.
X_poly = pd.DataFrame(X_poly, columns = poly.get_feature_names(housing_tr1.columns))
```

```python
#Feature scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X_poly)
housing_scaled_train = pd.DataFrame(scaled_data, columns = poly.get_feature_names(housing_tr1.columns))
```

```python
#check the shape to ensure rows are the same as the categorical variables to be concatenated later.
housing_scaled_train.shape
```

```python
housing_cat_train.shape
```

```python
#merge housing_scaled with housing_cat. Decided not to drop the extra columns (n-1)
#because of regularization https://stats.stackexchange.com/questions/231285/dropping-one-of-the-columns-when-using-one-hot-encoding
result_train = pd.concat([housing_scaled_train, housing_cat_train], axis=1, sort=False)
```

```python
#checking out the number of columns
result_train.head()
```

### 10. Modelling

Train-test-split training data then create X and y variables then fitting the model. Calculate rmse for the training-training dataset.

```python
#introduce train-test-split
X_train, X_test, y_train, y_test = train_test_split(result_train, train['saleprice'], test_size=0.2, random_state=42)
```

```python
#creating X and y variables for model
X = X_train
y = y_train
```

```python
#Model fitting for train data
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
model = lr.fit(X, y)
```

```python
#MSE calculation for train-train data
from sklearn.metrics import mean_squared_error
prediction = model.predict(X)
lin_mse = mean_squared_error(y, prediction)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
#check the score for the train-train set
score_train = model.score(X_train, y_train)
score_train
```

```python
#apply on the validation set which is the 'test' set that was split from the training dataset
score_test = model.score(X_test, y_test)
score_test
```

### 11. Regularization

Use Ridge to introduce penalty to the model and find the optimal alpha by adjusting the range of alphas used.

```python
#regularization using Ridge as Lasso seems too rigorous. Can try to set different alphas.
from sklearn.linear_model import Ridge, RidgeCV
r_alphas = np.linspace(0.01, 3, 100)
ridge_model = RidgeCV(alphas=r_alphas, store_cv_values=True)
ridge_model = ridge_model.fit(X, y)
# Here is the optimal value of alpha
ridge_optimal_alpha = ridge_model.alpha_
print(ridge_optimal_alpha)
ridge_opt = Ridge(alpha=ridge_optimal_alpha)
#cross_val_score(ridge_opt, X, y, cv=5).mean()
```

Alpha value is very low and this could be because the number of variables used is relatively low.

```python
#scores from regularization, does not seem to be better but might actually help in reducing overfit
''' This is if using cross_val_score.
ridge_cv_mean_mse = -cross_val_score(ridge_opt, X, y, cv=5, scoring='neg_mean_squared_error').mean()
ridge_cv_mean_r2 = cross_val_score(ridge_opt, X, y, cv=5, scoring='r2').mean()
ridge_cv_mean_mse, ridge_cv_mean_r2
'''
```

```python
#Ridge model optimized on the training-training set then fitted on the validation set, predicted and scored from the rmse.
ridge_opt = ridge_opt.fit(X_test, y_test)
prediction_ridge = ridge_opt.predict(X_test)
lin_mse = mean_squared_error(y_test, prediction_ridge)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
#lm plot
y_values_train = pd.DataFrame()
ridge_opt = ridge_opt.fit(result_train, train['saleprice'])
y_train_predicted1 = ridge_opt.predict(result_train)
y_values_train['predicted'] = y_train_predicted1
y_values_train['real'] = train['saleprice']
sns.lmplot(x='real', y='predicted', data=y_values_train)
ax = plt.gca()
ax.set_title('Lmplot for predicted saleprice with actual saleprice',fontsize=28);
```

Based on the lm plot, the ridge-optimized model that was obtained from training-training set fits well on the entire training set.
